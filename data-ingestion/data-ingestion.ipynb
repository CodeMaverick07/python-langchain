```python
#text loader 
# use langchain_community.document_loaders
from langchain_community.document_loaders import TextLoader
loader = TextLoader("data.txt")
print(loader)

#you need to load the data to create a useable data 

doc = loader.load()
type(doc[0])
```

```python
#loading pdf 
from langchain_community.document_loaders import PyPDFLoader
pdf = PyPDFLoader("java.pdf")
pdf = pdf.load()

```

```python
# web based loaders 
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader("https://myanimelist.net/anime/37450/Seishun_Buta_Yarou_wa_Bunny_Girl_Senpai_no_Yume_wo_Minai")
loader.load()
```

#Data transformation

```python
 # recersive text splitting
from langchain_text_splitters import RecursiveCharacterTextSplitter
 
split = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_doc = split.split_documents(pdf)
final_doc
```

```python
speed = ""
with open("data.txt") as d:
    speed = d.read()

speed = split.create_documents([speed])
speed
```
